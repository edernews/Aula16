```python
###  Passo 4 Dividir o DATASET para criar as base de treino  ###

x = iceCream[['Temperature']]
y = iceCream['Revenue']


```

```python
####### dividir a lista do DATAFRAME 

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25)
```

```python
x_train.shape
```

```python
#### Importar o Algoritmo LinearRegression da Biblioteca Sckitilean



from sklearn.linear_model import LinearRegression
```

```python
### esse comando gera a linha de tendencia ###

regressor = LinearRegression(fit_intercept=True)
regressor .fit(x_train, y_train)
```

```python
print(' o coeficiente Angular  (A)', regressor.coef_)
print(' O coeficiente Linear  (B)', regressor.intercept_)
```

```python
### Passo 6 executar o teste  ###

y_prevision = regressor.predict(x_test)

y_prevision
```
